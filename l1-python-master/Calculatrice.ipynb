```python
import tkinter as tk

def affiche_chiffre(chiffre) :
    global nb1, nb2, sym, affichage, text
    text += str(chiffre)
    affichage.set(text)
    if sym == "" :
        nb1 += str(chiffre)
    else :
        nb2 += str(chiffre)

def affiche_symbole(symbole) :
    global nb1, nb2, sym, affichage, text
    text +=(symbole)
    affichage.set(text)
    if sym != "" :
        nb1 = conversion(nb1)
        nb2 = conversion(nb2)
        if sym == "+" :
            res = nb1 + nb2
            text = str(res)
            affichage.set(text)
        elif sym == "-" :
            res = nb1 - nb2
            text = str(res)
            affichage.set(text)
        elif sym == "x" :
            res = nb1 * nb2
            text = str(res)
            affichage.set(text)
        elif sym == "รท" :
            res = nb1 / nb2
            text = str(res)
            affichage.set(text)
        elif sym == "%" :
            res = nb1 % nb2
            text = str(res)
            affichage.set(text)
        nb1=str(nb1)
        nb2=str(nb2)
    sym = symbole
    nb1 = text
    nb2= ""
        

def affiche_resultat() :
    global nb1, nb2, sym, affichage, text
    nb1 = conversion(nb1)
    nb2 = conversion(nb2)
    if sym == "+" :
        res = nb1 + nb2
        text = str(res)
        affichage.set(text)
    elif sym == "-" :
        res = nb1 - nb2
        text = str(res)
        affichage.set(text)
    elif sym == "x" :
        res = nb1 * nb2
        text = str(res)
        affichage.set(text)
    elif sym == "รท" :
        res = nb1 / nb2
        text = str(res)
        affichage.set(text)
    elif sym == "%" :
        res = nb1 % nb2
        text = str(res)
        affichage.set(text)
    nb1=str(nb1)
    nb2=str(nb2)
    
def conversion(n) :
    if n == "" :
        return 0
    else :
        return int(n)

cal = tk.Tk()

nb1 = ""
nb2 = ""
sym = ""
affichage = StringVar()
text = ""

cal.title("Calculatrice")
ecran = tk.Entry(cal, width=20, textvariable=affichage, bg="white")
ecran.grid(row=0, column= 0, rowspan=1, columnspan=4)
bouton1 = tk.Button(cal, text="1", command=lambda : affiche_chiffre(1), font = ("times", "20"))
bouton1.grid(row=1, column=1)
bouton2 = tk.Button(cal, text="2", command=lambda : affiche_chiffre(2), font = ("times", "20"))
bouton2.grid(row=1, column=2)
bouton3 = tk.Button(cal, text="3", command=lambda : affiche_chiffre(3), font = ("times", "20"))
bouton3.grid(row=1, column=3)
bouton4 = tk.Button(cal, text="4", command=lambda : affiche_chiffre(4), font = ("times", "20"))
bouton4.grid(row=2, column=1)
bouton5 = tk.Button(cal, text="5", command=lambda : affiche_chiffre(5), font = ("times", "20"))
bouton5.grid(row=2, column=2)
bouton6 = tk.Button(cal, text="6", command=lambda : affiche_chiffre(6), font = ("times", "20"))
bouton6.grid(row=2, column=3)
bouton7 = tk.Button(cal, text="7", command=lambda : affiche_chiffre(7), font = ("times", "20"))
bouton7.grid(row=3, column=1)
bouton8 = tk.Button(cal, text="8", command=lambda : affiche_chiffre(8), font = ("times", "20"))
bouton8.grid(row=3, column=2)
bouton9 = tk.Button(cal, text="9", command=lambda : affiche_chiffre(9), font = ("times", "20"))
bouton9.grid(row=3, column=3)
bouton0 = tk.Button(cal, text="0", command=lambda : affiche_chiffre(0), font = ("times", "20"))
bouton0.grid(row=4, column=2)
operation = tk.Button(cal, text="+", command=lambda : affiche_symbole("+"), font = ("times", "20"))
operation.grid(row=1, column=0)
soutraction = tk.Button(cal, text="-", command=lambda : affiche_symbole("-"), font = ("times", "20"))
soutraction.grid(row=2, column=0)
multiplication = tk.Button(cal, text="x", command=lambda : affiche_symbole("x"), font = ("times", "20"))
multiplication.grid(row=3, column=0)
division = tk.Button(cal, text="รท", command=lambda : affiche_symbole("รท"), font = ("times", "20"))
division.grid(row=4, column=0)
modulo = tk.Button(cal, text="%", command=lambda : affiche_symbole("%"), font = ("times", "20"))
modulo.grid(row=4, column=1)
egale = tk.Button(cal, text="=", command=lambda : affiche_resultat(), font = ("times", "20"))
egale.grid(row=4, column=3)

cal.mainloop()
```
