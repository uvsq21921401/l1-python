```python
###EXO N°1

import tkinter as tk

alter = 0
dec = 0 

def CRec(event) :
    global alter, dec
    alter = 1 - alter
    if 200 < event.x < 300 and 200 < event.y < 300 and dec == 0 :
        if alter == 0 :
            canvas.create_rectangle((200, 300), (300, 200), fill="red")
        else : 
            canvas.create_rectangle((200, 300), (300, 200), fill="blue")
    else :
        dec += 1

def Recommencer() :
    global dec
    dec = 0
    canvas.create_rectangle((200, 300), (300, 200), fill="red")

racine = tk.Tk()

racine.title("Exo n°1")
canvas = tk.Canvas(racine, bg="black", height=500, width=500, relief="groove")
canvas.grid()
bouton = tk.Button(racine, text="Recommencer", font = ("times", "30"), relief="groove", command  = Recommencer)
bouton.grid()
canvas.create_rectangle((200, 300), (300, 200), fill = "red")
clik = racine.bind("<Button-1>", CRec)

racine.mainloop()
```

```python
## EXO N°2

import tkinter as tk

dec = 0
x1 = -1
y1 = -1
cpt = 0
l1 = 0
l2 = 0

def Ligne(event) :
    global x1, y1, cpt, l1, l2, dec
    if dec != 0 :
        return
    cpt += 1
    print(cpt, x1, y1)
    if cpt == 1 or cpt == 3 :
        x1 = event.x
        y1 = event.y
    elif cpt == 5 :
        canvas.delete(l1)
        canvas.delete(l2)
        x1 = -1
        y1 = -1
        cpt = 0
    else :
        if cpt == 2 : 
            l1 = canvas.create_line((x1, y1), (event.x, event.y), fill="blue")
            x1 = -1
            y1 = -1
        else : 
            l2 = canvas.create_line((x1, y1), (event.x, event.y), fill="red")
            x1 = -1
            y1 = -1

def bouton() :
    global dec
    if dec == 0 :
        dec = 1
        bouton.configure(text="Restart")
    else :
        dec = 0
        bouton.configure(text="Pause")


racine = tk.Tk()

racine.title("Exo n°2")
canvas = tk.Canvas(racine, bg="white", height=500, width=500)
canvas.grid()
bouton = tk.Button(racine, text="Pause", font = ("times", "30"), relief="groove",  command=bouton)
bouton.grid()
canvas.bind("<Button-1>", Ligne)

racine.mainloop()
```

```python
## EXO N°3

import tkinter as tk

croix = 0
carre = 0
cercle = 0
forme = 0

def Affiche(event) :
    global croix, carre, cercle, forme
    if event.x < 166 and croix < 2 :
        croix += 1
        canvas.create_rectangle((event.x-25, event.y-25), (event.x+25, event.y+25), fill="cyan")
        forme += 1
        canvas.create_line((event.x-20, event.y-20), (event.x+20, event.y+20), fill="blue", width=10)
        forme += 1
        canvas.create_line((event.x-20, event.y+20), (event.x+20, event.y-20), fill="blue", width=10)
        forme += 1
    if 166 < event.x < 166*2 and carre < 3 :
        carre += 1
        forme += 1
        canvas.create_rectangle((event.x-25, event.y-25), (event.x+25, event.y+25), fill="green")
    if 166*2 < event.x and cercle < 3 :
        cercle += 1
        forme += 1
        canvas.create_oval((event.x-50, event.y-50), (event.x+50, event.y+50), fill="red")

def Redemarrer() :
    global forme, croix, carre, cercle
    for i in range(3, forme+3) :
        canvas.delete(i)
    croix = 0
    carre = 0
    cercle = 0

racine = tk.Tk()

racine.title("Exo n°3")
canvas= tk.Canvas(racine, bg="black", height=500, width=500)
canvas.grid()
bouton = tk.Button(racine, text="Redémarrer", font = ("times", "30"), relief="groove", command=Redemarrer)
bouton.grid()
canvas.create_line((166, 0), (166, 500), fill="white")
canvas.create_line((166*2, 0), (166*2, 500), fill="white")
canvas.bind("<Button-1>", Affiche)

racine.mainloop()
```

```python
## EXO N°4

import tkinter as tk

dec = 0
rayon = 25

def Taille(event) :
    global rayon, carre, dec
    if dec != 0 :
        return
    if 250-rayon < event.x < 250+rayon and 250-rayon < event.y < 250+rayon :
        if rayon < 10 :
            return
        else :
            canvas.delete(carre)
            rayon -=5
            carre = canvas.create_rectangle((250-rayon, 250+rayon), (250+rayon, 250-rayon), fill = "red", outline="red")
    else :
        if rayon > 50 :
            return
        else :
            canvas.delete(carre)
            rayon +=5
            carre = canvas.create_rectangle((250-rayon, 250+rayon), (250+rayon, 250-rayon), fill = "red", outline="red")

def bouton() :
    global dec
    if dec == 0 :
        bouton.configure(text="Restart")
        dec = 1
    else : 
        bouton.configure(text="Pause")
        dec = 0
    
racine = tk.Tk()

racine.title("Exo n°4")
canvas = tk.Canvas(racine, bg="white", height=500, width=500)
canvas.grid()
bouton = tk.Button(racine, text="Pause", font = ("times", "30"), relief="groove", command=bouton)
bouton.grid()
carre = canvas.create_rectangle((225, 275), (275, 225), fill = "red", outline="red")
canvas.bind("<Button-1>", Taille)

racine.mainloop()
```

```python
## EXON°5

import tkinter as tk

pst_inf = 200
pst_sup = 400
color_1 = "red"
color_2 = "blue"

def deplacement(event) :
    global pst_inf, pst_sup, l1, l2, color_1, color_2
    if color_1 == "red" and color_2 == "blue" :
        color_1, color_2 = color_2, color_1
    else :
        color_2, color_1 = color_1, color_2
    if pst_inf < event.x < pst_sup :
        canvas.delete(l1)
        canvas.delete(l2)
        pst_inf += 10
        pst_sup -= 10
        l1 = canvas.create_line((pst_inf, 0), (pst_inf, 600), fill=color_1, width=5)
        l2 = canvas.create_line((pst_sup, 0), (pst_sup, 600), fill=color_2, width=5)
    else :
        if event.x < pst_inf : 
            canvas.delete(l1)
            canvas.delete(l2)
            pst_inf -= 10
            pst_sup -= 10
            l1 = canvas.create_line((pst_inf, 0), (pst_inf, 600), fill=color_1, width=5)
            l2 = canvas.create_line((pst_sup, 0), (pst_sup, 600), fill=color_2, width=5)
        else :
            canvas.delete(l1)
            canvas.delete(l2)
            pst_inf += 10
            pst_sup += 10
            l1 = canvas.create_line((pst_inf, 0), (pst_inf, 600), fill=color_1, width=5)
            l2 = canvas.create_line((pst_sup, 0), (pst_sup, 600), fill=color_2, width=5)

def recommencer() : 
    global pst_inf, pst_sup, l1, l2, color_1, color_2
    canvas.delete(l1)
    canvas.delete(l2)
    pst_inf = 200
    pst_sup = 400
    color_1 = "red"
    color_2 = "blue"
    l1 = canvas.create_line((200, 0), (200, 600), fill=color_1, width=5)
    l2 = canvas.create_line((400, 0), (400, 600), fill=color_2, width=5)

racine = tk.Tk()

racine.title("Exo n°5")
canvas = tk.Canvas(racine, bg="white", height=600, width=600)
canvas.grid()
bouton = tk.Button(racine, text="Recommencer", font = ("times", "30"), relief="groove", command=recommencer)
bouton.grid()
l1 = canvas.create_line((200, 0), (200, 600), fill=color_1, width=5)
l2 = canvas.create_line((400, 0), (400, 600), fill=color_2, width=5)
canvas.bind("<Button-1>", deplacement)

racine.mainloop()
```

```python
## EXON°6

import tkinter as tk

l_color = []

def color(event) :
    global rond, l_color
    if 0 < event.x < 50 and 0 < event.y < 50 :
        canvas.itemconfigure(rond, fill = "green")
        l_color.append("green")
    elif 50 < event.x < 100 and 0 < event.y < 50 :
        canvas.itemconfigure(rond, fill = "yellow")
        l_color.append("yellow")
    elif 0 < event.x < 50 and 50 < event.y < 100 :
        canvas.itemconfigure(rond, fill = "blue")
        l_color.append("blue")
    else :
        canvas.itemconfigure(rond, fill = "black")
        l_color.append("black")

def annuler() :
    global rond, l_color
    if l_color == [] :
        return
    del l_color[-1]
    if l_color == [] :
        canvas.itemconfigure(rond, fill = "black")
    else :
        canvas.itemconfigure(rond, fill = l_color[-1])

racine = tk.Tk()

racine.title("Exo n°6")
canvas = tk.Canvas(racine, bg="white", height=500, width=500)
canvas.grid()
bouton = tk.Button(racine, text="Annuler", font = ("times", "30"), relief="groove", command=annuler)
bouton.grid(row=0, column=1)
canvas.create_rectangle((0, 0), (50, 50), fill = "green")
canvas.create_rectangle((50, 0), (100, 50), fill = "yellow")
canvas.create_rectangle((0, 50), (50, 100), fill = "blue")
rond = canvas.create_oval((200, 300), (300, 200), fill = "black")
canvas.bind("<Button-1>", color)

racine.mainloop()
```
