```python
###EXO N째1

import tkinter as tk

alter = 0
x = -1
y = -1

def CRec(event) :
    global alter, x, y
    x = event.x
    y = event.y
    alter = 1 - alter
    if 200 < x < 300 and 200 < y < 300 :
        if alter == 0 :
            canvas.create_rectangle((200, 300), (300, 200), fill="red")
        else : 
            canvas.create_rectangle((200, 300), (300, 200), fill="blue")
    else :
        clik.configure(state = "deactivate")

def Recommencer() :
    clik.configure(state = "activate")

racine = tk.Tk()

racine.title("Exo n째1")
canvas = tk.Canvas(racine, bg="black", height=500, width=500, relief="groove")
canvas.grid()
bouton = tk.Button(racine, text="Recommencer", font = ("times", "30"), relief="groove", command  = Recommencer)
bouton.grid()
canvas.create_rectangle((200, 300), (300, 200), fill = "red")
clik = racine.bind("<Button-1>", CRec)

racine.mainloop()
```

```python
## EXO N째2

import tkinter as tk

x1 = -1
y1 = -1
cpt = 0
l1 = 0
l2 = 0

def Ligne(event) :
    global x1, y1, cpt, l1, l2
    cpt +=1
    if x1 == -1 :
        x1 = event.x
        y1 = event.y
    else :
        if cpt == 2 : 
            l1 = canvas.create_line((x1, y1), (event.x, event.y), fill="blue")
            x1 = -1
            y1 = -1
        elif cpt == 4 : 
            l2 = canvas.create_line((x1, y1), (event.x, event.y), fill="red")
            x1 = -1
            y1 = -1
        elif cpt == 5 :
            canvas.delete(l1)
            canvas.delete(l2)
            cpt = 0

racine = tk.Tk()

racine.title("Exo n째2")
canvas = tk.Canvas(racine, bg="white", height=500, width=500)
canvas.grid()
bouton = tk.Button(racine, text="Pause", font = ("times", "30"), relief="groove")
bouton.grid()
clik = racine.bind("<Button-1>", Ligne)

racine.mainloop()
```
